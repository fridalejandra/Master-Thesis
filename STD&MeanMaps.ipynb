```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

fig = plt.figure(figsize=[12,10])

#May
ax = fig.add_subplot(231)
ax.set_title("May")
m = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
n = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
x, y = m(lon, lat)
x, y = n(lon, lat)
m.fillcontinents(color='white',lake_color='grey')
n.fillcontinents(color='white',lake_color='grey')
m.drawcoastlines()
n.drawcoastlines()
m.drawparallels(np.arange(-80.,0.,20.))
m.drawmeridians(np.arange(-180.,181.,20.))
m.drawmapboundary(fill_color='darkgrey')
n.drawparallels(np.arange(-80.,0.,20.))
n.drawmeridians(np.arange(-180.,181.,20.))
n.drawmapboundary(fill_color='darkgrey')
clr1=m.contourf(x,y,may,40,cmap="Blues",levels= [0,1,2,3,4,5])
clr2=n.contour(x, y, may1,20,levels=[0.5,1],cmap=plt.cm.jet,linewidths=1.7)
transform=ccrs.SouthPolarStereo()

#June
ax = fig.add_subplot(232)
ax.set_title("June")

j = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
j1 = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
x, y = j(lon, lat)
x, y = j1(lon, lat)
j.fillcontinents(color='white',lake_color='grey')
j1.fillcontinents(color='white',lake_color='grey')
j.drawcoastlines()
j1.drawcoastlines()
j.drawparallels(np.arange(-80.,0.,20.))
j.drawmeridians(np.arange(-180.,181.,20.))
j.drawmapboundary(fill_color='darkgrey')
j1.drawparallels(np.arange(-80.,0.,20.))
j1.drawmeridians(np.arange(-180.,181.,20.))
j1.drawmapboundary(fill_color='darkgrey')
j.contourf(x,y,jun,40,cmap="Blues",levels= [0,1,2,3,4,5])
j1.contour(x, y, jun1,20, levels=[0.5,1],cmap=plt.cm.jet,linewidths=1.7)
transform=ccrs.SouthPolarStereo()

#July
ax = fig.add_subplot(233)
ax.set_title("July")

jj = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
jj1 = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
x, y = jj(lon, lat)
x, y = jj1(lon, lat)
jj.fillcontinents(color='white',lake_color='grey')
jj1.fillcontinents(color='white',lake_color='grey')
jj.drawcoastlines()
jj1.drawcoastlines()
jj.drawparallels(np.arange(-80.,0.,20.))
jj.drawmeridians(np.arange(-180.,181.,20.))
jj.drawmapboundary(fill_color='darkgrey')
jj1.drawparallels(np.arange(-80.,0.,20.))
jj1.drawmeridians(np.arange(-180.,181.,20.))
jj1.drawmapboundary(fill_color='darkgrey')
jj.contourf(x,y,jul,40,cmap="Blues",levels= [0,1,2,3,4,5])
jj1.contour(x, y, jul1,20, levels=[0.5,1],cmap=plt.cm.jet,linewidths=1.7)
transform=ccrs.SouthPolarStereo()

#August
ax = fig.add_subplot(234)
ax.set_title("August")
a = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
a1 = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
x, y = a(lon, lat)
x, y = a1(lon, lat)
a.fillcontinents(color='white',lake_color='grey')
a1.fillcontinents(color='white',lake_color='grey')
a.drawcoastlines()
a1.drawcoastlines()
a.drawparallels(np.arange(-80.,0.,20.))
a.drawmeridians(np.arange(-180.,181.,20.))
a.drawmapboundary(fill_color='darkgrey')
a1.drawparallels(np.arange(-80.,0.,20.))
a1.drawmeridians(np.arange(-180.,181.,20.))
a1.drawmapboundary(fill_color='darkgrey')
a.contourf(x,y,aug,40,cmap="Blues",levels= [0,1,2,3,4,5])
a1.contour(x, y, aug1,20, levels=[0.5,1],cmap=plt.cm.jet,linewidths=1.7)
transform=ccrs.SouthPolarStereo()

#September
ax = fig.add_subplot(235)
ax.set_title("September")
s = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
s1 = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
x, y = s(lon, lat)
x, y = s1(lon, lat)
s.fillcontinents(color='white',lake_color='grey')
s1.fillcontinents(color='white',lake_color='grey')
s.drawcoastlines()
s1.drawcoastlines()
s.drawparallels(np.arange(-80.,0.,20.))
s.drawmeridians(np.arange(-180.,181.,20.))
s.drawmapboundary(fill_color='darkgrey')
s1.drawparallels(np.arange(-80.,0.,20.))
s1.drawmeridians(np.arange(-180.,181.,20.))
s1.drawmapboundary(fill_color='darkgrey')
s.contourf(x,y,sep,40,cmap="Blues",levels= [0,1,2,3,4,5])
s1.contour(x, y, sep1,20, levels=[0.5,1],cmap=plt.cm.jet,linewidths=1.7)
transform=ccrs.SouthPolarStereo()



#October
ax = fig.add_subplot(236)
ax.set_title("October")
oc = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
oc1 = Basemap(projection= 'spaeqd',boundinglat=-50,lon_0=-180,resolution='l')
x, y = oc(lon, lat)
x, y = oc1(lon, lat)
oc.fillcontinents(color='white',lake_color='grey')
oc1.fillcontinents(color='white',lake_color='grey')
oc.drawcoastlines()
oc1.drawcoastlines()
oc.drawparallels(np.arange(-80.,0.,20.))
oc.drawmeridians(np.arange(-180.,181.,20.))
oc.drawmapboundary(fill_color='darkgrey')
oc1.drawparallels(np.arange(-80.,0.,20.))
oc1.drawmeridians(np.arange(-180.,181.,20.))
oc1.drawmapboundary(fill_color='darkgrey')
oc.contourf(x,y,octo,40,cmap="Blues",levels= [0,1,2,3,4,5])
oc1.contour(x, y, octo1,20, levels=[0.5,1],cmap=plt.cm.jet,linewidths=1.7)
transform=ccrs.SouthPolarStereo()

fig.suptitle('Sea ice thickness climatologies 2002-2018', fontsize=15, y=1)
plt.subplots_adjust(wspace=0.05, hspace=0.05, right=0.8)
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
cbar_ax = fig.add_axes([-0.1, 0.15, 0.03, 0.7])
cbar_axx = fig.add_axes([1, 0.15, 0.03, 0.7])

fig.colorbar(clr2,cax=cbar_ax,label= 'SIT (m) standard deviation')
fig.colorbar(clr1, cax=cbar_axx,label= 'SIT (m) mean')

#plt.show(block = False)
#plt.savefig("climatologies_final.pdf")
plt.savefig("/Users/fridaperez/Developer/repos/local_repo/pub_plots/climatologies_sit_02-18.png",bbox_inches='tight',dpi=300)
plt.show(block = False)
```
